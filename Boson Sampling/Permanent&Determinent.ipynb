```python
import numpy as np             # NumPy is used for numerical operations and matrix manipulations
import itertools              # itertools provides tools for creating iterators for efficient looping
import time                   # time is used to measure the duration of computations
import matplotlib.pyplot as plt
```

```python
# Define a function to compute the determinant of a matrix using NumPy
def compute_determinant(matrix):
    return np.linalg.det(matrix)

# Define a function to compute the permanent of a matrix using a naive algorithm
def compute_permanent(matrix):
    n = matrix.shape[0]        # Get the number of rows (or columns) of the matrix
    total = 0                  # Initialize the total sum for the permanent
    # Loop over all possible permutations of row indices
    for perm in itertools.permutations(range(n)):
        prod = 1               # Initialize the product for the current permutation
        # Multiply the matrix elements according to the current permutation
        for i, j in enumerate(perm):
            prod *= matrix[i][j]
        total += prod         # Add the product to the total sum
    return total               # Return the computed permanent
```

```python
# Lists to store the results for varying matrix sizes
sizes = list(range(3, 12))
det_times = []
perm_times = []

# Compute determinant and permanent for matrix sizes from n=2 to n=7
for n in sizes:
    matrix = np.random.rand(n, n)
    
    # Time determinant computation
    start_time = time.time()
    compute_determinant(matrix)
    end_time = time.time()
    det_times.append(end_time - start_time)
    
    # Time permanent computation2
    start_time = time.time()
    compute_permanent(matrix)
    end_time = time.time()
    perm_times.append(end_time - start_time)
```

```python
# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, det_times, marker='o', label='Determinant Time', color='blue')
plt.plot(sizes, perm_times, marker='o', label='Permanent Time', color='red')
plt.xlabel('Matrix Size (n)')
plt.ylabel('Computation Time (seconds)')
#plt.yscale('log')
plt.title('Computation Time of Determinant vs Permanent')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
